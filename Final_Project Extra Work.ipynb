```python
import yfinance as yf
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
start_date = '2020-12-1' #1 December 2020
end_date = '2023-2-2'    #2 February 2023
```

```python
amazon = yf.download(tickers = "AMZN",
                  start = start_date,
                  end = end_date)
```

```python
amazon.head()
```

```python
amazon.reset_index(inplace=True)
amazon.head()
```

```python
plt.figure(figsize=(14,5))
sns.set_style("ticks")
sns.lineplot(data=amazon,x="Date",y='Close',color='blue')
sns.despine()
plt.title("The Stock Price of Amazon",size='x-large',color='red')
```
